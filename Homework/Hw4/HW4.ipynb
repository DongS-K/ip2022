```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img1 = cv2.imread('static/dsu1.jpg')
img2 = cv2.imread('static/dsu2.jpg')
rows,cols,ch = img1.shape
#img1 : [322,153],[760,212],[323,634],[768,586]
#img2 : [180,107],[821,69],[169,603],[819,628]
pts1 = np.float32([[322,153],[760,212],[323,634],[768,586]])
pts2 = np.float32([[180,107],[821,69],[169,603],[819,628]])
pts3 = np.float32([[0,0],[300,0],[0,300],[300,300]])
M = cv2.getPerspectiveTransform(pts1,pts3)
N = cv2.getPerspectiveTransform(pts2,pts3)
dst1 = cv2.warpPerspective(img1,M,(600,300))
dst2 = cv2.warpPerspective(img2,N,(300,300))
dst1[0:300, 300:600] = dst2
plt.subplot(111),plt.imshow(dst1),plt.title('Output')
plt.show()
```
