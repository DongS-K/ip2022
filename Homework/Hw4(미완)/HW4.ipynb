```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img1 = cv2.imread('static/dsu1.jpg')
img2 = cv2.imread('static/dsu2.jpg')
rows,cols,ch = img1.shape
#img1 : [322,153],[760,212],[323,634],[768,586]
#img2 : [180,107],[821,69],[169,603],[819,628]
pts1 = np.float32([[375,234],[888,213],[129,819],[845,861]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img1,M,(300,300))
plt.subplot(121),plt.imshow(img1),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
